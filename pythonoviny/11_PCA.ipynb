Nahodny proces ma 2 premenne $\mathbf{X}(\omega, t) = (X(\omega), Y(\omega))$

- Rozdelenim pravdepodobnosti $PDF(k) = Pr(X(\omega) = k) = Pr(X = k) = P_k$
- Distribucna funkcia $CDF(x) = Pr(X(\omega) < x) = \sum_{k < x}^{} P_k$
- $X(\omega) = Y(\omega) \Leftrightarrow CDF_X(t) = CDF_Y(t)$
- $\mathbf{X}(\omega, k), k = {1, 2}$
    - $CDF(t_1, t_2) = Pr(X_1 < t_1, X_2 < t_2)$
- $\mathbf{X}(\omega, t), k = {1, 2, \dots, n}$
    - $CDF_x(t_1, t_2, \dots, t_n) = Pr(X_1 < t_1, X_2 < t_2, \dots, X_n < t_n)$
    - Je rovnaky s $\mathbf{Y}(\omega, t)$ prave vtedy ked $CDF_X = CDF_Y$

Charakteristiky
- $EX$ - stredna hodnota
    - $E\mathbf{X}(\omega, t) = (EX_1, EX_2, \dots, EX_n) = \mathbf{m}(t)$
        - $t = {1, 2, \dots, n}$
- $DX$ - disperzia (rozptyl, variancia)
    - $D\mathbf{X}(\omega, t) = (DX_1, DX_2, \dots, DX_n) = \mathbf{d}(t)$
        - $t = {1, 2, \dots, n}$

Vycentrovany (stacionarny) nahodny proces
- $\mathbf{m}(t) = 0$
- $\mathbf{d}(t) = \sigma^2$

Kovariancia

Absolutne netusim co znamena hento vsetko, mam pocit ze to absolutne nesuvisi s tym co sa actually deje vo vypoctoch

Pome si dat vzorovy prikladik

Mame 5 datovych vektorov: $\begin{matrix} (0, 2) \\ (1, 0) \\ (2, 0) \\ (1, 1) \\ (1, 2) \end{matrix}$

Dame si ich do matice $\mathbb{X} = \begin{pmatrix} 0 & 2 \\ 1 & 0 \\ 2 & 0 \\ 1 & 1 \\ 1 & 2 \end{pmatrix}$

Je potrebne vycentrovat jednotlive stlpce matice

Najprv si zistime priemer jednotlivych stlpcov $\mathbb{m} = (1, 1)$

Potom odpocitame od stlpca prisludny priemer (v tomto pripade $-1$ pre kazdy stlpec) a dostaneme centrovanu maticu $\mathbb{X}_0 = \begin{pmatrix} -1 & 1 \\ 0 & -1 \\ 1 & -1 \\ 0 & 0 \\ 0 & 1 \end{pmatrix}$

Z tejto matice musime vypocitat kovariancnu maticu podla vzorca $\mathbb{R} = \dfrac{\mathbb{X}_0^T \cdot \mathbb{X}_0}{n - 1}$, kde $n = \text{pocet riadkov } \mathbb{X}_0$

$$\dfrac{
\begin{pmatrix} -1 & 0 & 1 & 0 & 0 \\ 1 & -1 & -1 & 0 & 1 \end{pmatrix} \cdot
\begin{pmatrix} -1 & 1 \\ 0 & -1 \\ 1 & -1 \\ 0 & 0 \\ 0 & 1 \end{pmatrix}
}{5 - 1} = \dfrac{
\begin{pmatrix} 2 & 2 \\ -2 & 4 \end{pmatrix}
}{4} = \begin{pmatrix} 0.5 & 0.5 \\ -0.5 & 1 \end{pmatrix}$$

Dalej musime vypocitat vlastne cisla matice

Hladame riesenie pre $(\mathbb{X}_0 - \lambda \mathbb{E}) = 0$, cize na diagonalu $\mathbb{X}_0$ dosadim $- \lambda$ a hladam take riesenie kde sa determinant tento matice bude rovnat $0$

Kedze je to matica $2 \times 2$, tak budeme riesit kvadraticku rovnicu a vysledkom budu 2 korene $=$ 2 vlastne cisla

$\begin{pmatrix} 0.5 - \lambda & -0.5 \\ -0.5 & 1 - \lambda \end{pmatrix}$

$\Rightarrow (0.5 - \lambda)(1 - \lambda) - 0.25 = 0$

$\Rightarrow 0.5 - 0.5\lambda -\lambda + \lambda^2 - 0.25 = 0$

$\Rightarrow \lambda^2 - 1.5\lambda + 0.25 = 0$

A podla vzorca kvadratickej rovnice $\dfrac{-b \pm \sqrt{b^2 - 4ac}}{2a}$ pocitame

$\dfrac{1.5 \pm \sqrt{1.5^2 - 4 \cdot 1 \cdot 0.25}}{2 \dot 1} = \dfrac{1.5 \pm 1.2}{2} \dot{=} \{ 1.35, 0.2 \}$

Vlastne cisla su korene tejto rovnice

$\lambda_1 = 1.35$

$\lambda_2 = 0.2$

Teraz ked mame vlastne cisla, potrebujeme si vypocitat aj vlastne vektory matice

Jednoducho dosadim $\lambda_1$ tam, kde sa nachadzalo $\lambda$ a pocitam

$\begin{pmatrix} 0.5 - 1.35 & 0.5 \\ -0.5 & 1 - 1.35 \end{pmatrix} \sim \begin{pmatrix} -0.85 & -0.5 \\ -0.5 & - 0.35 \end{pmatrix} \sim \begin{pmatrix} 1 & 0.59 \\ 1 & 0.7 \end{pmatrix}$

Tieto 2 riadky by mali byt linearne zavisle, ale chyba zaokruhlovania, takze...

$\begin{pmatrix} 1 & 0.59 \\ \cancel{1} & \cancel{0.7} \end{pmatrix} \rightarrow (1, 0.59)$

Povedzme ze prvy clen je $x$ a druhy clen je $y$

Z tohto mozeme zistit ze

$1x + 0.59y = 0 \Rightarrow x = -0.59y$

Ked si dosadime povedzme ze $y = 1$, tak dostaneme vlastny vektor $\mathbf{u}_1 = (-0.59, 1)$

Tento vektor ale musime vynormovat = podelit ho jeho dlzkou

$\mathbf{v_1} = \dfrac{\mathbf{u}_1}{|\mathbf{u}_1|} = \dfrac{(-0.59, 1)}{\sqrt{(-0.59)^2 + 1^2}} \dot{=} \dfrac{(-0.59, 1)}{1.16} = (-0.51, 0.86)$

$\mathbf{v_1} = (-0.51, 0.86)$

Posledny krok - postupne vynasobime kazdy riadok matice $\mathbb{X}$ s vlastnym vektorom $\mathbf{v_1}$ aby sme dostali jednotlive koeficienty

$$\begin{matrix}
c^1_1 = \langle \mathbb{x}_1, \mathbf{v_1}^T \rangle = (0 \cdot (-0.51) + 2 \cdot 0.86) = 1,72 \\
c^1_2 = \langle \mathbb{x}_2, \mathbf{v_1}^T \rangle = (1 \cdot (-0.51) + 0 \cdot 0.86) = -0.51 \\
c^1_3 = \langle \mathbb{x}_3, \mathbf{v_1}^T \rangle = (2 \cdot (-0.51) + 0 \cdot 0.86) = 1.02 \\
c^1_4 = \langle \mathbb{x}_4, \mathbf{v_1}^T \rangle = (1 \cdot (-0.51) + 1 \cdot 0.86) = 0.35 \\
c^1_5 = \langle \mathbb{x}_5, \mathbf{v_1}^T \rangle = (1 \cdot (-0.51) + 2 \cdot 0.86) = 1.21 \\
\end{matrix}$$

Este ak by sme chceli vediet mieru zachovania informacie, podelime pouzite vlastne cislo suctom vsetkych vlastnych cisel

$\dfrac{\lambda_1}{\sum \forall \lambda} = \dfrac{\lambda_1}{\lambda_1 + \lambda_2} = \dfrac{1.35}{1.35 + 0.2} \dot{=} 0.87 = 87 \%$

Cize pri tomto premietani sa zachova $87 \%$ povodnych informacii
