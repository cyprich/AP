# Harmonicka baza

Podoba sa mi to na mocniny (predchadzajuci subor na konci)

Behas po jednotkovej kruznici $n$-krat ($n$ = rozmer bazy)

Pri kazdom kroku sa posunies o $k$ kuskov ($k$ = cislo riadku, **! zacina sa od $0$**)

Je fess dobre si to predstavit ze co kde co robi kedy

**Harmonicka baza je ortogonalna**

$\mathbf{H_4} = \begin{pmatrix} 1 & 1 & 1 & 1 \\ 1 & i & -1 & -i \\ 1 & -1 & 1 & -1 \\ 1 & -i & -1 & i \end{pmatrix} = \begin{pmatrix} \mathbf{h_0} \\ \mathbf{h_1} \\ \mathbf{h_2} \\ \mathbf{h_3} \\ \end{pmatrix}$

$\mathbf{H_3} = \begin{pmatrix} 1 & 1 & 1 \\  1 & e^{i \dfrac{2 \pi}{3}} & e^{i \dfrac{4 \pi}{3}} \\ 1 & e^{i \dfrac{4 \pi}{3}} & e^{i \dfrac{2 \pi}{3}} \end{pmatrix} = \begin{pmatrix} 1 & 1 & 1 \\ 1 & - \dfrac{1}{2} + i \dfrac{\sqrt{3}}{2} & - \dfrac{1}{2} - i \dfrac{\sqrt{3}}{2}  \\ 1 & - \dfrac{1}{2} - i \dfrac{\sqrt{3}}{2} & - \dfrac{1}{2} + i \dfrac{\sqrt{3}}{2} \end{pmatrix} = \begin{pmatrix} \mathbf{h_0} \\ \mathbf{h_1} \\ \mathbf{h_2} \\ \end{pmatrix}$

$\mathbf{H_5} \approx \begin{pmatrix} 1 & 1 & 1 & 1 & 1 \\ 1 & 0.31 + i0.95 & -0.81 + i0.59 & -0.81 - i0.59 & 0.31 - i0.59 \\ 1 & -0.81 + i0.59 & 0.31 - i0.59 & 0.81 + i0.59 & -0.81 - i0.59 \\ 1 & -0.81 - i0.59 & 0.31 + i0.59 & 0.81 - i0.59 & -0.81 + i0.59 \\ 1 & 0.31 - i0.95 & -0.81 - i0.59 & -0.81 + i0.59 & 0.31 + i0.59 \\ \end{pmatrix}= \begin{pmatrix} \mathbf{h_0} \\ \mathbf{h_1} \\ \mathbf{h_2} \\ \mathbf{h_3} \\ \mathbf{h_4} \end{pmatrix}$

> Treba si vsimnut ze $\mathbf{h_1} = \mathbf{\overline{h_5}}$ a $\mathbf{h_2} = \mathbf{\overline{h_3}}$ a samozrejme ze sa to furt opakuje

### Vseobecny tvar $n$-teho vektora hamonickej bazy v $N$-rozmernom priestore

$\mathbf{h_n} = \left( \dots, e^{i \cdot \dfrac{2 \pi}{N} \cdot k } ,\dots \right)$, kde $k=\{0, 1, \dots, N-1 \}$ a $n=\{0, 1, \dots, N-1 \}$


### $k$-ta zlozka vektora

$2a_n \cdot cos \left( \dfrac{2 \pi}{N} \cdot nk \right) - 2b_n \cdot sin \left( \dfrac{2 \pi}{N} \cdot nk \right)$

Toto potom bude aj final funkcia ($y(t)$)


**Energia harmonickeho vektora = $N$**

$\langle \mathbf{h_n}, \mathbf{h_n} \rangle = N$


## Spektrum $\mathbf{c}$

Tiez oznacovane ako spektrum procesu

$\mathbf{c} = (c_0, c_1, \dots, c_{N-1})$

$c_0 = \dfrac{\langle \mathbf{f}, \mathbf{h_0} \rangle}{\langle \mathbf{h_0}, \mathbf{h_0} \rangle} = \dfrac{\langle \mathbf{f}, \mathbf{h_0} \rangle}{N}$

$c_1 = \dfrac{\langle \mathbf{f}, \mathbf{h_1} \rangle}{\langle \mathbf{h_1}, \mathbf{h_1} \rangle} = \dfrac{\langle \mathbf{f}, \mathbf{h_1} \rangle}{N}$

$c_2 = \dots$

**Nezabudnut ze druhy vektor v skalarnom sucine je komplexne zdruzeny!** $\Rightarrow \langle \mathbf{f}, \mathbf{h_0} \rangle = \mathbf{f} \cdot \mathbf{\overline{h_0}}$

$c_n = \overline{c}_{N-n}$




## Amplitudove spektrum $\mathbf{ca}$

Ktora vlna je ako zastupena

Velkosti jednotlivych hodnot spektra

$\mathbf{ca} = (|\mathbf{c_0}|, |\mathbf{c_1}|, \dots, |\mathbf{c_n}|) = (\sqrt{a_{c_0}^2, b_{c_0}^2}, \sqrt{a_{c_1}^2, b_{c_1}^2}, \dots)$

$\mathbf{ca_n} = \mathbf{\overline{ca_{N-n}}}$

**Pri parnych harmonickych funkciach treba strednu hodnotu amplitudoveho spektra podelit dvomi!!!**


## $fp = c_n \cdot \mathbf{h_n} + c_{N-n} \cdot \mathbf{h_{N-n}}$

Pri priemete treba dat aj komplexne zdruzeny




```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.pyplot import savefig
```

```python
f = np.array([2, 0, 3, 1, 4])
t = np.linspace(0, len(f)-1, len(f))
tt = np.linspace(0, len(f)-1, 1000)

# spektrum - funkcia fast fourier transform
c = sp.fft.fft(f)

# on spravi iba skalarny sucin <f, h> treba podelit energiou
c /= len(f)

# amplitudove spektrum
ca = abs(c)

# actually funkcia y(t)
y0 = c[0].real
y1 = 2 * c[1].real * np.cos((2 * np.pi/len(f)) * 1 * tt) - 2 * c[1].imag * np.sin((2 * np.pi/len(f)) * 1 * tt)
y2 = 2 * c[2].real * np.cos((2 * np.pi/len(f)) * 2 * tt) - 2 * c[2].imag * np.sin((2 * np.pi/len(f)) * 2 * tt)
y = y0 + y1 + y2

# priemety (to iste ako funkcia ale diskretne)
fp0 = c[0].real
fp1 = 2 * c[1].real * np.cos((2 * np.pi/len(f)) * 1 * t) - 2 * c[1].imag * np.sin((2 * np.pi/len(f)) * 1 * t)
fp2 = 2 * c[2].real * np.cos((2 * np.pi/len(f)) * 2 * t) - 2 * c[2].imag * np.sin((2 * np.pi/len(f)) * 2 * t)

# vypisovacky, vykreslovacky
print(fp1)
df = pd.DataFrame([t, f, fp1, fp2, (fp0 + fp1 + fp2)], index=["t", "f", "fp1", "fp2", "fp"])
# df = pd.DataFrame([t, f], index=["t", "f"])
display(df)

df = pd.DataFrame([i for i in c], index=[f"c{i}" for i in range(len(c))])
display(df)

df = pd.DataFrame([i for i in ca], index=[f"ca{i}" for i in range(len(ca))])
display(df)


plt.figure()
plt.plot(f, label="f")
plt.title("Povodne data")
plt.legend()

plt.figure()
plt.plot(ca, label="ca")
plt.title("Amplitudove spektrum")
plt.legend()

plt.figure()
plt.plot(f, label="f")
plt.plot(tt, (y2 + y0), label="y2")
plt.title("Funkcia $y_2$ (najvyznamnejsia)")
plt.legend()
savefig("obrazky/08_1.png")

plt.figure()
plt.plot(f, label="f")
plt.plot(tt, (y1 + y0), label="y1")
plt.title("Funkcia $y_1$")
plt.legend()
savefig("obrazky/08_2.png")

plt.figure()
plt.plot(f, "o-", label="f")
plt.plot(tt, y, label="y")
plt.title("Funkcia $y$ (interpolacia)")
plt.legend()
savefig("obrazky/08_3.png")


```

```python
# najdite amplitudove spektrum procesu f
f = np.array([1, 0, 4, 0, 4, 3])
t = np.linspace(0, len(f)-1, len(f))
# tt = np.linspace(0, len(f)-1, 1000)

c = sp.fft.fft(f)
c /= len(f)

# amplitudove spektrum
ca = abs(c)
if int(len(f) % 2) == 0:
    middle = int(len(f) / 2)
    ca[middle] /= 2

df = pd.DataFrame({"c": c, "ca": [round(i, 4) for i in ca.real]})
display(df)

plt.figure()
plt.plot(f, label="$f$")
plt.title("Nahodny proces $f$")
plt.legend()

plt.figure()
plt.plot(ca, label="$ca$")
plt.title("Amplitudove spektrum nahodneho procesu $f$")
_ = plt.legend()
```
